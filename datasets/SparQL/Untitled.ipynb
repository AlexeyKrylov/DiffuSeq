```python
with open('spec-english_train_split.txt', 'r') as f:
    spec = f.read()

with open('src-english_train_split.txt', 'r', encoding='utf8') as f:
    src = f.readlines()
```

```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
```

```python
Counter(spec.replace('\n', ' ').split(' '))
```

```python
len(dict(Counter(src.lower().replace('\n', ' ').replace('?', '').replace('.', '').split()).most_common()).keys())
```

```python
s = set(list(dict(Counter(src.lower().replace('\n', ' ').replace('?', '').split(' ')).most_common()).keys()) + list(dict(Counter(spec.replace('\n', ' ').split(' ')).most_common()).keys()))
```

```python
tokenizer_1 = Tokenizer(BPE(unk_token="[UNK]"))
trainer = BpeTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]"], vocab_size=5000)
tokenizer_1.pre_tokenizer = Whitespace()
tokenizer_1.train_from_iterator([i.strip() for i in src], trainer)
print(len(tokenizer_1.get_vocab()))
```
